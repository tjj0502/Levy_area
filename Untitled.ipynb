```python
import torch
import signatory
import numpy as np
from torch import nn
import math
import Utilities, Augmentations
from scipy.stats import norm
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
class Path_Generator:
    def __init__(self, batches, length, dim):
        self.batches = batches
        self.dim = dim
        self.length = length


class BM_Path_Generator(Path_Generator):
    def __init__(self, batches, length, dim):
        super(BM_Path_Generator, self).__init__(batches, length, dim)

    def generate(self):
        BM_paths = torch.empty(self.batches, self.length+1, self.dim)
        nn.init.normal_(BM_paths, 0, math.sqrt(1 / self.length))
        BM_paths[:, 0, :] *= 0
        BM_paths = torch.cumsum(BM_paths, dim=1)
        return BM_paths
    
def bridge_var(t, t_start = 0., t_end = 0.01):
    return torch.tensor((t_end-t)*(t)/(t_end - t_start)).repeat(2)

def bridge_mean(x_start, x_end, t, t_start = 0., t_end = 0.01):
    return x_start + (t - t_start)*(x_end - x_start)/(t_end - t_start)
```

```python
def bridge_cov(s, t, t_start = 0., t_end = 0.01):
    '''
    s must be smaller that t
    '''
    assert s<=t, "s must be smaller that t"
    # if (t_end-t)*(s-t_start)/(t_end - t_start) < 0:
    #     print(s, t, t_start, t_end, (t_end-t)*(s-t_start)/(t_end - t_start))
    #     return
    return torch.tensor((t_end-t)*(s-t_start)/(t_end - t_start))

def bridge_var_matrix(n, t_start = 0., t_end = 0.01):
    matrix = torch.zeros([n,n])
    for i in range(1, n+1):
        for j in range(i, n+1):
            s = t_start + i/(n+1) * (t_end - t_start)
            t = t_start+ j/(n+1) * (t_end - t_start)
            matrix[i-1,j-1] = bridge_cov(s, t, t_start, t_end)
            matrix[j-1,i-1] = bridge_cov(s, t, t_start, t_end)
    return matrix
 
def bridge_mean_t(x_start, x_end, n, t_start = 0., t_end = 0.01):
    mean = torch.zeros(n)
    for i in range(1,n+1):
        t = t_start + i/(n+1) * (t_end - t_start)
        mean[i-1] = x_start + (t - t_start)*(x_end - x_start)/(t_end - t_start)
    return mean
```

```python
batch_size = 10000
n = 1000
dim = 2
idx = 0
length = 100
n_paths = 5
level = 2
torch.manual_seed(52)
bms_generator = BM_Path_Generator(batches = batch_size, length = length, dim = 2)
bms = bms_generator.generate()
middle_points = torch.zeros([batch_size, n, dim])
for i in range(dim):
    bridge_mean = bridge_mean_t(bms[idx,0,i], bms[idx,50,i], n, t_start = 0., t_end = 1)
    bridge_var = bridge_var_matrix(n, t_start = 0., t_end = 1)
    middle_points_gen = torch.distributions.MultivariateNormal(bridge_mean, bridge_var)
    middle_points[:,:,i] = middle_points_gen.sample([batch_size,]).reshape([batch_size, -1])
init = bms[idx,[0,-1]].unsqueeze(0).repeat((batch_size,1,1))
inserted_path = torch.cat([init[:,0,:].unsqueeze(1), middle_points, init[:,1,:].unsqueeze(1)], axis = 1)
time_dix = torch.linspace(0, 1, inserted_path.shape[1]).reshape(1, -1, 1).repeat(batch_size, 1, 1)
aug_inserted_path = torch.cat([time_dix, inserted_path], dim=-1)
```

```python
def sum_A(dim, coeffs = None):
    use_coeffs = coeffs is not None
    if use_coeffs:
        assert dim*(dim-1)/2 ==  torch.numel(coeffs), "Dimension does not agree"
        coeffs = torch.flatten(coeffs)
    A = np.zeros([dim,dim])
    k=0
    for i in range(dim):
        for j in range(i+1, dim):
            if use_coeffs:
                A[i,j] = -1*coeffs[k]
                k+=1
            else:
                A[i,j] = -1
            A[j,i] = -A[i,j]
    return torch.tensor(A)


def torch_decomposition(A):
    assert (torch.norm(A+A.t()) < 1e-5), "A is not skew-symmetric"
    assert (torch.norm(A) > 1e-10), "A is zero matrix"
    dim = A.shape[0]
    n = dim//2
    evals, evecs = torch.linalg.eig(A)
    evals = -evals # Make sure -lambda i comes before than lambda i
    V = evecs.conj().clone() # Change the eigenvectors correspondently
    
    # print(evals)
    # print(V)
    
    # torch.abs(evals.imag)
    _, keys = torch.sort(torch.abs(evals.imag), descending = True)
    evals = evals[keys]
    V = V[:,keys]
    # Sort eigenvalues by the norm of the imaginary part
    # print(evals)
    # print(V)
    
    S = torch.diag(evals)
    W_real = torch.tensor([[0.,1.],[1.,0.]])
    W_img = torch.tensor([[1.,0.],[0.,1.]])
    W = 1/torch.sqrt(torch.tensor(2)) * torch.complex(W_real, W_img).to(dtype = S.dtype)
    
    # Initialize the matrices Omega and Sigma
    Omega = W.clone()
    J = torch.tensor([[0.,1.],[-1.,0.]]).to(dtype = S.dtype)
    Sigma = torch.norm(evals[0]) * J.clone()
    
    # We need to normalize those eigenvectors with imaginary eigenvalues
    # V = evecs.conj().clone()
    for i in range(n-1):
        print(evals[2*(i+1)], torch.norm(evals[2*(i+1)]))
        Sigma = torch.block_diag(Sigma, torch.norm(evals[2*(i+1)])*J)
        # If imaginary, rescale it by 1/sqrt(-i)
        if evals[2*i].imag != 0.:
            # print('need to be normalized')
            V[:,[2*i,2*i+1]] = 1/torch.sqrt(torch.tensor(-1j)) * V[:,[2*i,2*i+1]]
        if evals[2*(i+1)].imag != 0.:    
            Omega = torch.block_diag(Omega,W)
        else:
            Omega = torch.block_diag(Omega,torch.eye(2).to(dtype = S.dtype))
           
    # V = 1/torch.sqrt(torch.tensor(-1j)) * evecs.conj()
    # V[:,-1] = torch.sqrt(torch.tensor(-1j)) * V[:,-1]

        
    # i = n-1
    if evals[2*(n-1)].imag != 0.:
        # print('need to be normalized')
        V[:,[2*(n-1),2*(n-1)+1]] = 1/torch.sqrt(torch.tensor(-1j)) * V[:,[2*(n-1),2*(n-1)+1]]
        
    # print('after scaling')
    # print(V)
    if dim%2 == 1:
        Omega = torch.block_diag(Omega, torch.eye(1).to(dtype = S.dtype))
        Sigma = torch.block_diag(Sigma, torch.zeros(1).to(dtype = S.dtype))
    # print('V after scaling')
    # print(V)
    # print('Omega')
    # print(Omega)
    # print('Sigma')
    # print(Sigma)
    # print(S.shape)
    # print(torch.norm(Omega @ S @ Omega.conj().t() - Sigma))

    U = V @ Omega.conj().t()
    
    
    # Evec relation:
    i = 0
    # Check eigen decomposition
    assert (torch.norm(A.to(dtype = S.dtype) @ V[:,i] - (-evals[i]) * (-V[:,i])) < 1e-5)
    assert (torch.norm(A.to(dtype = S.dtype) @ V[:,i].conj() - (-evals[i]) * V[:,i].conj()) < 1e-5)
    assert (torch.norm(V @ S @ V.conj().t() - A.to(dtype = S.dtype)) < 1e-5)
    
    
    # Check Omega S Omega^+ = Sigma
    assert (torch.norm(Omega @ S @ Omega.conj().t() - Sigma) < 1e-5), Omega
    
    # Check Omega^+ Sigma Omega = S
    assert (torch.norm(Omega.conj().t() @ Sigma @ Omega - S) < 1e-5)
    
    # Check U^T A U = Sigma
    assert (torch.norm(U.t() @ A.to(dtype = S.dtype) @ U - Sigma) < 1e-5)
    
    # Check that U is orthogonal
    assert (torch.norm(U @ U.t() - torch.eye(dim)) < 1e-5), U @ U.t()
    assert (torch.norm(U.t() @ U - torch.eye(dim)) < 1e-5), U.t() @ U
    # Check that U is real
    assert (torch.norm(U - U.conj()) < 1e-5)
    
    print(evals.imag)
    
    return U
```
